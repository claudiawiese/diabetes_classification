### Basic Analysis 

```python
sns.countplot(x ='Diabetes', data = df)
```

As we can see we have much more patients without diabetes than with diabetes. In other words we have an imbalanced dataset and certainly need to deal with imbalance when constructing our classification modesl with techniques such as random over or undersampling.

Let's check the distribution of the other variables.

```python
columns_to_plot = df.columns[df.columns != 'Diabetes']

# Set the color palette
palette = sns.color_palette("husl", len(columns_to_plot))

# Create a figure for subplots
num_vars = len(columns_to_plot)
n_cols = 3  # Number of columns
n_rows = (num_vars + n_cols - 1) // n_cols  # Calculate rows needed

fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 5 * n_rows))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop through the columns and plot each distribution
for i, (ax, column, color) in enumerate(zip(axes, columns_to_plot, palette)):
    sns.histplot(df[column], ax=ax, kde=True, color=color)
    ax.set_title(f'Distribution of {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')

# Hide any unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The distributions suggest typical variability expected in a population with a range of metabolic profiles, from healthy to those with significant insulin resistance or diabetes.

The right skewness observed in insulin, BMI, and diabetes pedigree function highlights the presence of outliers, particularly in insulin levels, which are indicative of severe metabolic disturbances in some individuals.

These outliers in insulin are biologically plausible given the accompanying skewness in BMI and glucose distributions, suggesting the presence of a subgroup with severe metabolic dysregulation, aligning with clinical expectations in a diverse population sample.

However there seems to be a problem with some of the BMI values. The BMI cannot be equal to 0. 

### Check for outliers and extreme values

```python
df.describe()
```

The BMI can not be 0. Let's replace these values with the median BMI.
Also a PlasmaGlucoseConcerntration equal to 0 is not possible. We also will replace these values with the median value. Same for the DiastolicBloodPressue which also cannot be 0. 

```python
# Function to replace zero values with median for specific columns
def replace_zeros_with_median(df, columns):
    median_replaced_df = df.copy()
    for col in columns:
        median_value = median_replaced_df[col].median()
        median_replaced_df[col] = median_replaced_df[col].replace(0, median_value)
    return median_replaced_df

# Columns to replace zero values with their median
columns_to_replace_zeros = ['PlasmaGlucoseConcentration', 'DiastolicBloodPressure', 'BMI']

# Applying the replacement
df_cleaned = replace_zeros_with_median(df, columns_to_replace_zeros)

# Displaying the updated dataset for review
df_cleaned.head()
```

```python
df_cleaned = df[df['BMI'] != 0]
df_cleaned = df_cleaned[df_cleaned['PlasmaGlucoseConcentration'] != 0]
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(data=df_cleaned)
plt.xticks(rotation=45)
plt.title("Boxplot to Identify Outliers")
plt.show()
```

We have a lot of outliers for SerumInsulin (2-Hour serum insulin (mu U/ml)). Let's check if all these values are reasonable. 

```python
#Checking for values above 500
df_cleaned[df_cleaned.SerumInsulin > 500]
```

The extremely high insulin levels in these cases are biologically plausible given the context of high glucose levels, elevated BMIs, and, in some cases, strong family history (diabetes pedigree function). These values represent severe insulin resistance often seen in morbid obesity, advanced type 2 diabetes, and individuals with strong genetic predispositions to diabetes. However, some of the higher values, especially those nearing or exceeding 800 μU/mL, are exceptionally rare and, while plausible, warrant careful clinical validation to ensure accuracy and rule out potential lab errors.

So let's do the analysis with these outliers and if the results are too distorted we can still replace the outliers with median values for example.

### Normalizing the Data

Last but not least we might want to normalize our variables because they are on different scales (eg. age vs SerumInsulin).

Standard scaling ensures that each variable contributes equally to the correlation calculation, eliminating the dominance of variables with larger ranges

```python
# Define the features to be scaled
features = ['Pregnancies', 'PlasmaGlucoseConcentration', 'DiastolicBloodPressure', 
            'TricepsSkinFoldThickness', 'SerumInsulin', 'BMI', 
            'DiabetesPedigreeFunction', 'Age']

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the selected features
df_scaled = scaler.fit_transform(df[features])

# Convert the scaled data back to a DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=features)

# Add the target variable 'Diabetes' back to the DataFrame
df_scaled['Diabetes'] = df['Diabetes']

# Display the first few rows of the scaled DataFrame
df_scaled.head()
```

### Correlation Matrix 

```python
correlation_matrix = df_scaled.corr()
correlation_matrix.style.background_gradient(cmap='coolwarm')
```

The two variables that have the strongest relationship with Diabetes are PlasmaGlucoseConcentration with a correlation coefficient of 0.47 and the Body Mass Index (BMI) with a coefficient of 0.29. Other significant positive correlation include age (0.24) and  pregnancies (0.22). 

Based on the correlation matrix, none of the explanatory variables exhibit correlations high enough (above 0.7 or 0.8) to suggest severe multicollinearity. This implies that your model should not suffer from the negative impacts of multicollinearity, such as inflated standard errors and unreliable coefficient estimates.
